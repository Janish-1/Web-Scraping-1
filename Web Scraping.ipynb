```python
# For Fairfax County Public Schools
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
import chromedriver_autoinstaller as chromedriver
import pandas as pd
import os
chromedriver.install()

# Invoking webdriver
driver = webdriver.Chrome()# Adding the chromedriverpath here 
try:
    driver.get("https://fairfaxhs.fcps.edu/staff-directory")
    rows =  len(driver.find_elements_by_xpath('//table/tbody/tr'))
    columns = len(driver.find_elements_by_xpath('//table/thead/tr/th'))
    print("Rows:",rows,"Columns:",columns)
    Headers = []
    heads = driver.find_elements_by_xpath('//table/thead/tr/th')
    for i in heads:
        Headers.append(i.text)
    driver.close()
except:
    print("Invalid url")
    
print("Headers:",Headers[:-1])

x = 0
names = []
positions = []
try:
    while(x < 1000):
        driver1 = webdriver.Chrome()
        driver1.get("https://fairfaxhs.fcps.edu/staff-directory?field_last_name_from=&field_last_name_to=&items_per_page=10&keywords=&page="+str(x)
        )
        for i in range(1,rows+1):
            # Fetching the Names
            Finder1 = driver1.find_element_by_xpath('//table/tbody/tr['+str(i)+']/td[1]/a[2]').text # Second Name Fetch
            Finder1_1 = driver1.find_element_by_xpath('//table/tbody/tr['+str(i)+']/td[1]/a[1]').text # First Name Fetch
            name = (Finder1_1+" "+Finder1) # Combining the Names
            names.append(name)

            #Fetching the Position
            Finder2 = driver1.find_element_by_xpath('//table/tbody/tr['+str(i)+']/td[2]').text # Position Fetch
            positions.append(Finder2)
        x += 1
        # Closing the webdriver
        driver1.close()
except:
    print("Pages End")
    
for k in range(len(names)):
    print(k," ",names[k]," ",positions[k])
```

## Note For the Above Scraping
1. First step was to cycle the pages So i used while loop with end condition
2. For the Rows and Columns i used Selenium to get the total count of that particular tags by using xpath
3. Then i used a for loop to get the data of all the names avaliable on the webpage,doing this i had successfully scraped the whole website

```python
# For Miami Dade County Public Schools
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
import chromedriver_autoinstaller as chromedriver
import pandas as pd
import time

chromedriver.install()
# Invoking Webdriver
x = 0
driver = webdriver.Chrome()
driver.get("https://www3.dadeschools.net/EmployeeSearch/location")
time.sleep(15)
print("Wait Over")
rows = len(driver.find_elements_by_xpath('//table/tbody/tr'))
columns = driver.find_elements_by_xpath('//table/thead/tr/th')
print("Rows: ",rows,"Columns: ",len(columns))
for i in columns:
    if i == 0:
        print("Columns: ","S.No.",end=" ")
    else:
        print(i.text,end = " ")
print("")
df = pd.DataFrame(columns=['S.No','Location Number','Location Name','Location Address','Phone/Fax Number','Region Center'])
s1,s2,s3,s4,s5=[],[],[],[],[]
try:
    for i in range(1,rows):
        loc_number = driver.find_element_by_xpath('//table/tbody/tr['+str(i)+']/td[1]/a[@href]')
        loc_number1 = loc_number.get_attribute('href')
        loc_name = driver.find_element_by_xpath('//table/tbody/tr['+str(i)+']/td[2]').text
        loc_address = driver.find_element_by_xpath('//table/tbody/tr['+str(i)+']/td[3]').text
        ph_fax_num = driver.find_element_by_xpath('//table/tbody/tr['+str(i)+']/td[4]').text
        region_center = driver.find_element_by_xpath('//table/tbody/tr['+str(i)+']/td[5]').text
        s1.append(loc_number1)
        s2.append(loc_name)
        s3.append(loc_address)
        s4.append(ph_fax_num)
        s5.append(region_center)
except:
    print("Pages Ended")
driver.close()
```
